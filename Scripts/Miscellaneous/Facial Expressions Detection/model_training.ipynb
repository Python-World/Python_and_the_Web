# Imports

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

# Data Preparation using Data Augmentation

```python
train_data_path = "./Data/train"
test_data_path = "./Data/test"
```

```python
train_data_generator = ImageDataGenerator(rescale=1./255,
                                          rotation_range=20,
                                          shear_range=0.2,
                                          zoom_range=0.2,
                                          horizontal_flip=True)
```

```python
test_data_generator = ImageDataGenerator(rescale=1./255)
```

```python
training_data = train_data_generator.flow_from_directory(directory=train_data_path, 
                                                         target_size=(48, 48),
                                                         color_mode='grayscale',
                                                         class_mode='categorical',
                                                         batch_size=64)
```

```python
testing_data = test_data_generator.flow_from_directory(directory=test_data_path, 
                                                        target_size=(48, 48),
                                                        color_mode='grayscale',
                                                        class_mode='categorical',
                                                        batch_size=64)
```

```python
training_data.class_indices
```

# Model Building

```python
model = Sequential()
```

```python
model.add(Conv2D(32, 3, input_shape=(48, 48, 1), activation='relu', padding='same'))
model.add(MaxPooling2D((2, 2), strides=2))
model.add(Conv2D(64, 3, activation='relu', padding='same'))
model.add(Dropout(0.2))
model.add(MaxPooling2D((2, 2), strides=2))
model.add(Conv2D(64, 3, activation='relu', padding='same'))
model.add(Dropout(0.2))
model.add(MaxPooling2D((2, 2), strides=2))
model.add(Conv2D(128, 3, activation='relu', padding='same'))
model.add(Flatten())
model.add(Dense(4, activation='softmax'))
```

```python
model.summary()
```

# Compiling the Model

```python
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

# Model Training

```python
model.fit_generator(training_data, validation_data=testing_data, epochs=300, steps_per_epoch=len(training_data), 
                    validation_steps=len(testing_data), verbose=1)
```

# Saving the model

```python
model.save("detection_model.h5")
```
